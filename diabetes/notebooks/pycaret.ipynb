```python
from pycaret.datasets import get_data
from pycaret.classification import setup
import pandas as pd 
```

```python
df = pd.read_csv('../data/01_raw/diabetes.csv')
```

```python
df.columns
```

```python
model_setup = setup(data=df, target="Outcome")
```

```python
best_model = model_setup.compare_models()
```

```python
output_df = model_setup.predict_model(best_model, data=df)
```

```python
output_df.loc[:, ["Outcome", "prediction_label"]]
```

```python
model_setup.evaluate_model(best_model)
```

```python
pipeline, model_path = model_setup.save_model(best_model, "my-best-model")
```
